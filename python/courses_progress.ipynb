## Transripts

```python
def parse_transcript(F, name, replace = False):
    global req
    df =  [[
        a.split()[0]+"_"+a.split()[1], a.split()[-2]
    ] for a in F.splitlines() if len(a)>0]
    df = pd.DataFrame(df)
    df.columns = ["Code", name]
    df.set_index("Code", inplace = True)
    new_req = pd.concat([req, df], axis=1)
    if(replace):
        req = new_req.copy()
    return new_req
```
