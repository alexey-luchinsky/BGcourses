## Transripts

```python
def parse_transcript(F, season):
    df =  [{
        "Code":aa.split()[0]+"_"+aa.split()[1], 
        "Name":" ".join(aa.split()[2:-4]), 
        "Credits":int(float(aa.split()[-4])),
        "season":season,
        "grade":aa.split()[-2]} for aa in F.splitlines() if len(aa)>3]
    df = pd.DataFrame(df)
    # df.columns = ["Code", name]

    return df
```
